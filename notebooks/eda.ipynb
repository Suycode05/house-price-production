```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 150)
pd.set_option('display.max_rows',150)

df=pd.read_csv('./data/train.csv')
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
missing=df.isna().mean()*100
missing=missing[missing>0].sort_values(ascending=False)
missing
```

```python
plt.figure(figsize=(10,8))
plt.bar(missing.index,missing)
plt.title('Percentage missing by column')
plt.ylabel('%Missing')
plt.xticks(rotation=90)
plt.show()
```

```python
numerical=df.select_dtypes(include=['int64','float64']).columns
numerical
```

```python
df[numerical].hist(figsize=(20,18),bins=40)
plt.suptitle('Numerical Feature Distributions')
plt.show()
```

```python
df[numerical].skew().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df['SalePrice'])
plt.title('SalePrice Outliers')
plt.show()
```

```python
important_features=['OverallQual','GrLivArea','TotalBsmtSF','GarageArea']
for i in important_features:
    plt.figure(figsize=(10,8))
    sns.boxplot(df[i])
    plt.title(f'{i} Outliers')
    plt.show()
#
```

```python
categorical=df.select_dtypes(include=['object']).columns
categorical
```

```python
for i in categorical[:6]:
  plt.figure(figsize=(10,8))
  sns.countplot(data=df,x=i)
  plt.xticks(rotation=45)
  plt.title(f"Count plot of {i}")
  plt.show()
```

```python
correlation=df.corr(numeric_only=True)
correlation
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation,cmap='coolwarm',annot=False)
plt.title('Correlation Matrix')
plt.show()
```

```python
corr_target=correlation['SalePrice'].sort_values(ascending=False)
corr_target
```

```python
for i in ['GrLivArea','TotalBsmtSF','1stFlrSF']:
  plt.figure(figsize=(6,4))
  sns.scatterplot(x=df[i],y=df['SalePrice'])
  plt.title(f'SalePrice vs {i}')
  plt.show()
```

```python
for i in ['OverallQual','Neighborhood','HouseStyle']:
  plt.figure(figsize=(12,5))
  sns.boxplot(x=df[i],y=df['SalePrice'])
  plt.xticks(rotation=90)
  plt.title(f'SalePrice vs {i}')
  plt.show()
```
